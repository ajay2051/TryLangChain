```python
from openai import embeddings

# Continuation of 03

dataset = {}
for i, embedding in enumerate(embeddings):
    dataset[i] = embedding

import os

from upstash_vector import Index, Vector

index = Index(url=os.environ.get("UPSTASH_VECTOR_URL"), token=os.environ.get("UPSTASH_VECTOR_TOKEN"))


vectors = []
for key, value in dataset.items():
    print(key, value)
    my_id = key
    vectors.append(Vector(id=my_id, vector=value)) 
    
index.upsert(
  vectors=vectors
)
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
from openai import OpenAI
client = OpenAI(api_key=OPENAI_API_KEY)
def get_embedding(text, model="text-embedding-3-small"):
    text=text.replace("\n", " ")
    return client.embeddings.create(input=[text], model=model)

query_str = "hello world"
query_embedding = get_embedding(query_str)

results = index.query(
  vector=query_embedding,
  top_k=1,
  include_vectors=True,
  include_metadata=True
)
for result in results:
    print(result.id, result.score, result.metadata)
```
