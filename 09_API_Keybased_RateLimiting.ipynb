```python
import os
from dotenv import load_dotenv
load_dotenv()

API_ACCESS_KEY = os.environ.get("API_ACCESS_KEY")
```

```python
import requests
base_url = 'https://127.0.0.1:8000/'
data = {"input": "Tell me about new york"}
r = requests.get(base_url, json=data, headers={"X-API-KEY": f"{API_ACCESS_KEY}"})
r.status_code
```

```python
chain_endpoint = f"{base_url}chain/"
from langserve import RemoteRunnable
runner = RemoteRunnable(chain_endpoint)
runner.invoke("What is NY")
```
